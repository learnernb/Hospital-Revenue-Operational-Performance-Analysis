```python
import matplotlib.pyplot as plt

months = list(range(1, 13))
revenue = [
    3300246, 3187167, 3609171, 3411084,
    3598923, 3443135, 3561860, 3387516,
    3433736, 3462545, 3539029, 3631596
]

# Convert to Crores for cleaner display
revenue_cr = [r/10000000 for r in revenue]

plt.figure(figsize=(10,6))
plt.plot(months, revenue_cr, marker='o')

plt.xlabel("Month")
plt.ylabel("Revenue ( Crore)")
plt.title("Monthly Revenue Trend - 2025")

plt.grid(True)

# Highlight peak and lowest month
max_rev = max(revenue_cr)
min_rev = min(revenue_cr)

plt.scatter(months[revenue_cr.index(max_rev)], max_rev)
plt.scatter(months[revenue_cr.index(min_rev)], min_rev)

plt.show()
```

```python
import matplotlib.pyplot as plt

doctors = [
    "Amy Patel",
    "Michael Salazar",
    "Lindsey Reid",
    "Julie Garrett",
    "Louis Gomez"
]

revenues = [
    5026263,
    4962621,
    4958925,
    4954486,
    4685057
]

# Convert to Lakhs
revenues_lakh = [r / 100000 for r in revenues]

plt.figure(figsize=(10,6))
bars = plt.bar(doctors, revenues_lakh)

plt.xlabel("Doctor")
plt.ylabel("Revenue ( Lakh)")
plt.title("Top 5 Revenue Generating Doctors - 2025")

plt.xticks(rotation=30)
plt.grid(axis='y')

# Add value labels on top
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             f"{height:.1f}",
             ha='center',
             va='bottom')

plt.show()
```
